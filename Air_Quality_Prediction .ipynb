```python
# importing the necessary libraries
import os
import requests
from bs4 import BeautifulSoup as bs
import csv
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

# web scraping and data collection


```python
#Retreiving the html code from the website for every month between the years 2013-2018

def get_html(start_year,end_year):
    for year in range(start_year,end_year+1):
        for month in range(1,13):
            if month<10:
                url=f'https://en.tutiempo.net/climate/0{month}-{year}/ws-421820.html'
            else:
                url=f'https://en.tutiempo.net/climate/{month}-{year}/ws-421820.html'
                
            data=requests.get(url).text
            
            if not os.path.exists(f'Data/html_data/{year}'):
                os.makedirs(f'Data/html_data/{year}')
                
            with open(f'Data/html_data/{year}/{month}.html','w') as f:
                f.write(data)
                
```

```python
#calling the function
get_html(2013,2015)
```

```python
#Retreiving the independent variables from the html files

def html_data(month,year):
    with open(f'Data/html_data/{year}/{month}.html','r') as f:
        data=f.read()
    
    temp=[]
    final=[]
    
    tbl_data=bs(data)
    
    for table in tbl_data.find_all('table',attrs={'class':'medias mensuales numspan'}):       
        for tbody in table:
            for tr in tbody:
                    row_data=tr.get_text()
                    temp.append(row_data)
                    
    num=len(temp)/15
    for t in range(round(num)):
        new_temp = []
        for i in range(15):
            new_temp.append(temp[0])
            temp.pop(0)
        final.append(new_temp)
        
    final.pop(len(final)-1)
    final.pop(0)
    
    for a in range(len(final)):
        final[a].pop(6)
        final[a].pop(13)
        final[a].pop(12)
        final[a].pop(11)
        final[a].pop(10)
        final[a].pop(9)  
        final[a].pop(0)
        
        
    return final

#Here the empty columns are removed 
```

The website has only the independent variables and the dependent variable 'PM2.5' has been retrieved using a web API and stored as a csv file. 
The website has average data for each day but the dependent variable from the API has been recorded for every hour
Therefore we need to find the average PM2.5 for each day
An user defined module named year_data has been created for the same purpose which calculates the average PM2.5 for every day for each year separately 


```python
from year_data import avg_PM_2013, avg_PM_2014, avg_PM_2015
```

```python
#writing all the data into a CSV 
if not os.path.exists('Data/real_data'):
    os.makedirs('Data/real_data')

for year in range(2013,2016):
    final_data=[]
    with open(f'Data/real_data/real{year}.csv','w') as f:
        
        header=['T','TM','Tm','SLP','H','VV','V','VM','PM 2.5']
        writer=csv.writer(f,dialect='excel')
        writer.writerow(header)
        
    for month in range(1,13):
        temp=html_data(month,year)
        final_data+=temp
        
    pm=getattr(sys.modules[__name__],f'avg_PM_{year}')()
    
    for i in range(len(final_data)-1):
        final_data[i].insert(8,pm[i])
        
    with open(f'Data/real_data/real{year}.csv','a') as f:
        writer=csv.writer(f,dialect='excel')
        for row in final_data:
            empty=0
            for item in row:
                if item=="" or item=='-':
                    empty=1
            if empty!=1:
                writer.writerow(row)
            
        
```

```python
#combining all the data together
def final(start,end):
    df_list=[]
    final=[]
    for year in range(start,end):
        df=pd.read_csv(f'Data/real_data/real{year}.csv')
        df_list.append(df)
        final=pd.concat(df_list,ignore_index=True)
    return final
```

```python
final(2013,2016)
```

```python
final_csv=final(2013,2016)
```

```python
#saving the final data
final_csv.to_csv('Final_data.csv')
```

# Data preprocessing

```python
#removing unwanted column(s)
df=df.iloc[:,1:]
df.head()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#removing the records with missing values
df=df.dropna()
df.isnull().sum()
```

```python
#handling outliers
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1
df=df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR))).any(axis=1)]
```

```python
df.info()
```

```python
#splitting the data into dependent and independent variables
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#feature importance
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
feature_imp=model.feature_importances_
```

```python
pd.DataFrame(feature_imp,index=x.columns,columns=['Feature Importance'])
```

```python
#Train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=101)
```

# model building

### linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(xtrain,ytrain)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
#predictions
y_predict_test=LR.predict(xtest)
y_predict_train=LR.predict(xtrain)
```

```python
#Loss function
from sklearn import metrics
```

```python
print('Training:')
print('MSE:',metrics.mean_squared_error(ytrain,y_predict_train))
print('RMSE:',np.sqrt(metrics.mean_squared_error(ytrain,y_predict_train)))
print('MAE:',metrics.mean_absolute_error(ytrain,y_predict_train))
print('\n')
print('Testing:')
print('MSE:',metrics.mean_squared_error(ytest,y_predict_test))
print('RMSE:',np.sqrt(metrics.mean_squared_error(ytest,y_predict_test)))
print('MAE:',metrics.mean_absolute_error(ytest,y_predict_test))
```

```python
#Coefficient of determination(r2 score)
print('Training:')
print(LR.score(xtrain,ytrain))
print('\n')
print('Testing:')
print(LR.score(xtest,ytest))
```

The random forest regressor too came out to be an overfit model.
To reduce the overfitting,we can do hyperparameter tuning

```python
#Hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
#parameters to be tuned
n_estimators=[500,700,900,1100,1300]
max_depth=[9,10,11,12,13,14]
max_features=['auto','sqrt']
min_samples_split=[2,4,6,8,10]
min_samples_leaf=[2,4,6,8,10]
max_leaf_nodes=[11,13,15,18,20,23]

```

```python
param={'n_estimators':n_estimators,'max_depth':max_depth,'max_features':max_features,'min_samples_split':min_samples_split,'min_samples_leaf':min_samples_leaf,
      'max_leaf_nodes':max_leaf_nodes}
```

```python
HyperParams=RandomizedSearchCV(rf,param,cv=10,scoring='neg_mean_squared_error',verbose=2)
```

```python
HyperParams.fit(xtrain,ytrain)
```

```python
HyperParams.best_params_
```

```python
rf=RandomForestRegressor(n_estimators= 1300,max_depth=12, max_features='auto', min_samples_split=4,min_samples_leaf=2,
                         max_leaf_nodes=13,bootstrap=True,oob_score=True)
```

```python
rf.fit(xtrain,ytrain)
```

```python
#Coefficient of determination(r2 score)
print('Training:')
print(rf.score(xtrain,ytrain))
print('\n')
print('Testing:')
print(rf.score(xtest,ytest))
```
